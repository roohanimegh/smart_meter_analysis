```python
#helper functions and main function to convert pandas dataframes to pyspark dataframes (easier to work with)
sc = SparkContext("local","App Name")
sql = SQLContext(sc)

# Auxiliar functions
def equivalent_type(f):
    
    tz = pytz.timezone('America/Los Angeles')
    
    if f == 'datetime64[ns]': return DateType()
    elif f == 'int64': return LongType()
    elif f == 'int32': return IntegerType()
    elif f == 'float64': return FloatType()
    else: return StringType()

def define_structure(string, format_type):
    try: typo = equivalent_type(format_type)
    except: typo = StringType()
    return StructField(string, typo)


# Given pandas dataframe, it will return a spark's dataframe.
def pandas_to_spark(pandas_df):
    columns = pandas_df.columns.tolist()
    types = pandas_df.dtypes.tolist()
    struct_list = []
    for column, typo in zip(columns, types): 
      struct_list.append(define_structure(column, typo))
    p_schema = StructType(struct_list)
    return sql.createDataFrame(pandas_df, p_schema)
```

To assume this test is fit for an independent group t-test, the variances per day of the datasets should be similar betwee the two groups.

Lets look at daily variance between the standard meter users and the Time of User meter users 

```python
daily_acorn_df.head()
```

```python
daily_variance = daily_acorn_df.groupby(['stdortou','day'])['energy_mean'].var()
daily_variance = daily_variance.to_frame()
daily_variance.reset_index(inplace = True)

daily_variance_std = daily_variance[daily_variance['stdortou']=='Std']
daily_variance_tou = daily_variance[daily_variance['stdortou']=='ToU']

daily_variance.head()
```

```python
fig, ax1 = plt.subplots(figsize = (20,5))
ax1.plot(daily_variance_std.day, daily_variance_std.energy_mean, label = 'standard', color = 'orange')
ax1.plot(daily_variance_tou.day, daily_variance_tou.energy_mean, label = 'Time of User', color = 'pink')
ax1.set_ylabel('variation of energy reading')
ax1.legend() 

plt.title('variance of mean energy readings per day')
fig.tight_layout()
plt.show()
```

The Standard meter users show a high variance every day this trial was run. This makes sense as those users seem to feel more freedom to adjust their usage more frequently throughout the day with an equal tarriff.

The set tarrifs for smart meter users may further restrict the amount those HHs would change their energy usage throughout the day. This will be lower the daily variance as users are tempted to lower their energy usage during high tarriff times. 

```python
daily_sd = daily_acorn_df.groupby(['stdortou','day'])['energy_mean'].std()
daily_sd = daily_sd.to_frame()
daily_sd.reset_index(inplace = True)

daily_sd.head()
```

```python
daily_sd_std = daily_sd[daily_sd['stdortou']=='Std']
daily_sd_tou = daily_sd[daily_sd['stdortou']=='ToU']

#now join this on the day 

daily_sd_comparison = daily_sd_std.merge(daily_sd_tou, on = ['day'], how = 'left')
daily_sd_comparison['diff_in_sd'] = (daily_sd_comparison['energy_mean_x'] - daily_sd_comparison['energy_mean_y'])
def f(row):
    if row['diff_in_sd'] >= row[2]:
        val = 'not_equal'
    else:
        val = 'equal'
    return val

daily_sd_comparison['equal_or_not'] = daily_sd_comparison.apply(f, axis=1)

daily_sd_comparison.groupby('equal_or_not').count().head()
```

This VERY rough test shows that the differences are similar enough to ASSUME EQUAL SD/VARIANCE. 

The daily variance plot above further illustrates this as a small variance can be seen daily in the duration of this trial. 

Even with the uneveness of sample size, we can go ahead and carry out an independent t-test without being concerned for it producing invalid results (higher type 1 error rate, this being that the test assumes a difference in energy usage even though there isn't).

## T test comparing Standard Meters to Smart Meters

The T - test is a powerful statistical test which can help establish whether there is a statistical difference within two independent groups of data. We know that these groups are independent of eachother (i.e one HH could have either used a smart meter or a standard meter in the duration of the trial, not both!). To be able to carry out these tests the data set should roughly follow a normal distribution and the variance should be similar. 

As energy consumption is largely correlated with weather (which can be seen with the plot above). It seems that energy consumption is largely dictated by the daily temperature, and this feature is the same for every HH in London. With this being the case, I assume that most houses will largely use the same amount of energy so the energy means will be closely related to the overall daily mean and hence follow a normal distribution. 

Another assumption made to carry out valid independent t- tests are that the two groups have a similar variance. I have proved earlier on that the variances are similar, therefore we can continue with this test. 

```python
daily_std_w_temp = daily_std_df.merge(weather, on = ['day'], how = 'left')
daily_tou_w_temp = daily_tou_df.merge(weather, on = ['day'], how = 'left')

#we will be using max temp for plotting, convert to farenheit so it can be better interpreted by american audiences

daily_std_w_temp['max_temp_frht'] = (daily_std_w_temp['max_temp']*(9/5))+32
daily_tou_w_temp['max_temp_frht'] = (daily_tou_w_temp['max_temp']*(9/5))+32

daily_std_w_temp.head()
```

```python
daily_std_w_temp.sort_values("day", inplace=True)
day_list = daily_std_w_temp['day'].unique()
len(day_list)
```

```python
df = pd.DataFrame()
final_result= pd.DataFrame()

for day in day_list:
    
    df_std = daily_std_w_temp[daily_std_w_temp['day']==day]
    df_tou = daily_tou_w_temp[daily_tou_w_temp['day']==day]

    t_test_result = {ttest_ind(df_std.dropna()['energy_mean'],df_tou.dropna()['energy_mean'])}
    result_df = pd.DataFrame(t_test_result, columns =['tvalue','pvalue'])

    if df is None:
        final_result = df 
    else:
        final_result = pd.concat([final_result,result_df])

final_result.head()
```

```python
final_result.reset_index(drop = True, inplace = True)
# final_result.drop(['level_0','index'], axis = 1, inplace = True)
final_result.head()
```

```python
def stat_diff(row):
    if row['tvalue'] >= row['pvalue']:
        val = 1
    else:
        val = 0
    return val

ind_test_overall['tvalue'] = ind_test_overall['tvalue'].abs()
ind_test_overall['diff']  = ind_test_overall.apply(stat_diff, axis = 1)

ind_test_overall.head()
```

```python
#NULL HYPOTHESIS = TWO ENERGY MEANS ARE THE SAME 
#ALT HYPOTHESUS = TWO ENERGY MEANS ARE NOT THE SAME 

#THEREFORE IF T VALUE >= P VALUE, THEN REJECT HULL HYPOTHESIS (STD METER AND TOU METER READINGS ARE NOT THE SAME)
#IF P VALUE >= T VALUE, THEN FAIL TO REJECT HULL HYPOTHESIS (STD METER AND TOU METER READINGS ARE THE SAME)

#NB 'diff' == 1 <-- different means 
#NB 'diff' == 0 <-- same means 

ind_test_overall.dropna(how='any' ,inplace = True)
ind_test_overall.groupby('diff').count()
```

```python
#p tests for accorn groups 

df = pd.DataFrame()
final_result= pd.DataFrame()

def ind_t_test(dfstd,dftou):
    
    df = pd.DataFrame()
    final_result= pd.DataFrame()
    
#     daily_std_w_temp.sort_values("day", inplace=True)
    day_list = daily_std_w_temp['day'].unique()

    for day in day_list:
    
        df_std = dfstd[dfstd['day']==day]
        df_tou = dftou[dftou['day']==day]

        t_test_result = {ttest_ind(df_std.dropna()['energy_mean'],df_tou.dropna()['energy_mean'])}
        result_df = pd.DataFrame(t_test_result, columns =['tvalue','pvalue'])

        if df is None:
            final_result = df 
        else:
            final_result = pd.concat([final_result,result_df])

    final_result.reset_index(drop = True, inplace = True)
    day_df = pd.DataFrame(day_list, columns = ['day'])
    ind_test_overall = pd.merge(day_df, final_result, left_index=True, right_index=True)
    
    ind_test_overall['tvalue'] = ind_test_overall['tvalue'].abs()
    ind_test_overall['diff']  = ind_test_overall.apply(stat_diff, axis = 1)
    
    return ind_test_overall
```
